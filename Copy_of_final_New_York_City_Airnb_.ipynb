```python
# Calculate the quartiles, interquartile range, and outlier limits based on the original "price" column
q1 = a["price"].quantile(0.25)
q3 = a["price"].quantile(0.75)
iqr = q3 - q1
upper_limit_price = q3 + 1.5 * iqr
lower_limit_price = q1 - 1.5 * iqr

# Identify outliers
outliers_price = a[(a["price"] > upper_limit_price) | (a["price"] < lower_limit_price)]

# Remove outliers
a = a[(a["price"] >= lower_limit_price) & (a["price"] <= upper_limit_price)]
print(upper_limit_price)
print(lower_limit_price)
```

```python
room_type=a["room_type"].value_counts()
plt.pie(x = room_type.values, labels = room_type.index, autopct= '%1.2f%%')
plt.title("Room Type")
```

```python
# Donut chart
neighbourhood_groups = np.unique(a['neighbourhood_group'])
neighbourhood_group_counts = a["neighbourhood_group"].value_counts()
plt.pie(x = neighbourhood_group_counts.values, labels = neighbourhood_group_counts.index, autopct = "%1.2f%%")
circle = plt.Circle((0, 0), 0.3, color='white')
plt.gca().add_artist(circle)
plt.title("Neighbourhood Group Distribution")
```
