# MEAN OF ALL NUMERICALS VALUE

```python
# mean of all numerical value
sales =np.mean(a["Sales"])
print("sales_mean:", sales)
discount = np.mean(a["Discount"])
print("discount_mean:", discount)
profit = np.mean(a["Profit"])
print("profit_mean:", profit)
```

# MODE OF ALL NUMERICAL VALUE

```python
# mode
sales =statistics.mode(a["Sales"])
print("sales_mode:", sales)
discount = statistics.mean(a["Discount"])
print("discount_mode:", discount)
profit = statistics.mean(a["Profit"])
print("profit_mode:", profit)
```

MEADIAN OF ALL NUMERICAL VALUE

```python
#median
sales =np.median(a["Sales"])
print("sales_median:", sales)
discount = np.median(a["Discount"])
print("discount_median:", discount)
profit = np.median(a["Profit"])
print("profit_median:", profit)
```

#STD OF ALL NUMERICAL VALUE

```python
#std deviation
sales =np.std(a["Sales"])
print("sales_std:", sales)
discount = np.std(a["Discount"])
print("discount_std:", discount)
profit = np.std(a["Profit"])
print("profit_std:", profit)
```

# PIE CHART

```python
region_counts = a["Region"].value_counts()
cn = region_counts.index
cv = region_counts.values
plt.pie(x=cv, labels=cn, autopct="%1.2f%%")
plt.title('Distribution of Regions')
plt.show()
```

```python
a.columns
```

# DONUT CHART

```python
category_counts = a['Category'].value_counts()

# Plot the pie chart as a donut chart
plt.pie(x=category_counts.values, labels=category_counts.index, autopct='%1.2f%%')

circle = plt.Circle((0,0), 0.3, color='white')

plt.gca().add_artist(circle)

plt.title('Distribution of Categories')

```
